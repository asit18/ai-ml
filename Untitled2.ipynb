```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import math # Added math import


```

```python
def create_sample_data() -> tuple[np.ndarray, np.ndarray]:
    """Generates sample data for a simple regression problem."""
    np.random.seed(42)
    X = np.random.rand(100, 1) * 10
    y = 2 * X**2 + 3 * X + 5 + np.random.randn(100, 1) * 20
    return X, y
def get_data_frame(file_path: str) -> pd.DataFrame:
    """Generates a sample data frame for a simple regression problem."""
    df = pd.read_csv(file_path)
    df.head()
    df.describe()
    df.info()
    df.shape
    df.columns
    return df

def train_polynomial_regression(X: np.ndarray, y: np.ndarray, degree: int = 2) -> Pipeline:
    """Trains a polynomial regression model."""
    print(f"\nTraining Polynomial Regression Model with degree {degree}...")
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X, y)
    print("Model training complete.")
    return model

def evaluate_model(model: Pipeline, X: np.ndarray, y: np.ndarray) -> float:
    """Evaluates the model and prints the Mean Squared Error."""
    print("\nEvaluating Model...")
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    print(f"Mean Squared Error: {mse:.2f}")
    return mse

def main() -> None:
    """Main function to orchestrate the ML workflow."""
    print("Starting Python ML script...")

    # 1. Generate Data
    df = get_data_frame("sample_data/california_housing_train.csv")
    X = df.values[:, :-1]
    y = df.values[:, -1]
    # Get original feature names
    feature_names = df.columns[:-1]
    print(f"Generated {len(X)} data points.")

    # 2. Train Model
    poly_reg_model = train_polynomial_regression(X, y, degree=2)

    # Get polynomial feature names
    poly_features_step = poly_reg_model.named_steps['polynomialfeatures']
    transformed_feature_names = poly_features_step.get_feature_names_out(feature_names)
    print(f"Generated {len(transformed_feature_names)} polynomial features.")
    print("Polynomial Features:")
    for name in transformed_feature_names:
        print(f"  - {name}")
    # Get coefficients
    coefficients = poly_reg_model.named_steps['linearregression'].coef_

    print("\nCoefficients for each polynomial feature:")
    for name, coef in zip(transformed_feature_names, coefficients):
        print(f"  - {name}: {coef:.2f}")

    # 3. Evaluate Model
    mse_value = evaluate_model(poly_reg_model, X, y)

    # Example of using various math functions with scalar values
    if mse_value > 0:
        print(f"  - Square root of MSE: {math.sqrt(mse_value):.2f}")
        print(f"  - Natural logarithm of MSE: {math.log(mse_value):.2f}")
        # print(f"  - Exponent (e^MSE/1000): {math.exp(mse_value / 1000):.2f}") # Divided to keep value reasonable
    else:
        print("MSE is not positive, cannot calculate log or square root.")

    abs_mse = math.fabs(mse_value)
    print(f"  - Absolute value of MSE: {abs_mse:.2f}")

    ceil_mse = math.ceil(mse_value)
    floor_mse = math.floor(mse_value)
    print(f"  - Ceiling of MSE: {ceil_mse}")
    print(f"  - Floor of MSE: {floor_mse}")

    print("Python ML script finished.")

# This block ensures `main()` only runs when the script is executed directly.
if __name__ == '__main__':
    main()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Re-run data loading and model training to get the necessary variables
df = get_data_frame("sample_data/california_housing_train.csv")
X = df.values[:, :-1]
y = df.values[:, -1]

# Train the model
poly_reg_model = train_polynomial_regression(X, y, degree=2)

# Make predictions
y_pred = poly_reg_model.predict(X)
```

```python
# Create a scatter plot of actual vs. predicted values
fig = plt.figure(figsize=(10, 8))
sns.scatterplot(x=y, y=y_pred, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2) # Add a diagonal line for perfect prediction
plt.xlabel('Actual House Values')
plt.ylabel('Predicted House Values')
plt.title('Actual vs. Predicted House Values')
plt.grid(True)
plt.show()
```
